```python
df_zero_mask = df == 0
# To examine every row and from the second column
# elements that begin with 0
# using the all method along the column axis (axis=1)
feature_zero_mask = df_zero_mask.iloc[:,1:].all(axis=1)
sum(feature_zero_mask)
```

```python
# Creating One Hot Encoding
# Create an empty column for categorical labels

df_clean_1['EDUCATION_CAT'] = 'none'

# Examine the first few rows of the DataFrame
df_clean_1[['EDUCATION', 'EDUCATION_CAT']].head(10)
```

```python
# Map the numbers in EDUCATION to the strings they represent
# firstly create a dictionary mapping the education categories

cat_mapping = {
    1: "graduate school",
    2: "university",
    3: "high school",
    4: "others"
}

df_clean_1['EDUCATION_CAT'] = df_clean_1['EDUCATION'].map(cat_mapping)
df_clean_1[['EDUCATION', 'EDUCATION_CAT']].head(10)
```

```python
# one hot encoded columns are called dummy variables
# The result is a new dataframe
edu_ohe = pd.get_dummies(df_clean_1['EDUCATION_CAT'])
edu_ohe.head(10)
```

```python
# Concatenating one-hot encoded Dataframe to
# the original dataframe

df_with_ohe = pd.concat([df_clean_1, edu_ohe], axis=1)
df_with_ohe[['EDUCATION_CAT', 'graduate school',
             'high school', 'university', 'others']].head(10)
```
